```python
import cv2

# Create a VideoCapture object to access the camera
cap = cv2.VideoCapture(0)  # 0 represents the default camera

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Failed to open the camera")
    exit()

# Read and display frames from the camera
while True:
    ret, frame = cap.read()

    # If frame is not available, exit the loop
    if not ret:
        break

    # Rotate the frame by 180 degrees
    rotated_frame = cv2.rotate(frame, cv2.ROTATE_180)


    # Display the rotated frame in a window named "Camera Feed"
    cv2.imshow("Camera Feed", rotated_frame)

    # Wait for the 'q' key to be pressed to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import RPi.GPIO as GPIO
import time
import pigpio
import pygame

ESC_PIN = 4
servo_pin = 18

GPIO.setmode(GPIO.BCM)
GPIO.setup(ESC_PIN, GPIO.OUT)
GPIO.setup(servo_pin, GPIO.OUT)

pwm_esc = GPIO.PWM(ESC_PIN, 50)
pwm_esc.start(0)

servo_pwm = GPIO.PWM(servo_pin, 50)
servo_pwm.start(2.5)

def set_speed(speed):
    duty_cycle = speed / 100 * 10 + 5
    pwm_esc.ChangeDutyCycle(duty_cycle)

def set_servo_position(angle):
    pulse_width = (angle / 18) + 2
    servo_pwm.ChangeDutyCycle(pulse_width)

# Initialisation de Pygame et du joystick
pygame.init()
pygame.joystick.init()

try:
    print("Calibrating ESC....")
    set_speed(0)
    time.sleep(2)
    set_speed(100)
    time.sleep(2)
    set_speed(0)
    time.sleep(2)

    joystick = pygame.joystick.Joystick(0)
    joystick.init()

    print("Joystick initialized: ", joystick.get_name())

    while True:
        for event in pygame.event.get():
            if event.type == pygame.JOYAXISMOTION:
                # Contrôle de la vitesse avec l'axe Y du joystick
                speed = int(joystick.get_axis(1) * 100)  # Ajustez l'index de l'axe selon votre joystick
                set_speed(speed)

                # Contrôle de l'angle du servo avec l'axe X du joystick
                angle = int((joystick.get_axis(0) + 1) * 90)  # Ajustez l'index de l'axe selon votre joystick
                set_servo_position(angle)

            if event.type == pygame.QUIT:
                raise KeyboardInterrupt

except KeyboardInterrupt:
    print("Stopped by user")

finally:
    pwm_esc.stop()
    servo_pwm.stop()
    GPIO.cleanup()
    pygame.quit()
```

```python
import cv2
import mediapipe as mp
from picamera2 import Picamera2
import time




class poseDetector():
    def __init__(self,mode=False, upBody = False, smooth = True, detectionCon=0.5, trackCon=0.5):
        self.mode = mode
        self.upBody = upBody
        self.smooth = smooth
        self.detectionCon = detectionCon
        self.trackCon = trackCon

        self.mpDraw = mp.solutions.drawing_utils
        self.mpPose = mp.solutions.pose
        self.pose = self.mpPose.Pose(
            static_image_mode= self.mode,
            smooth_landmarks= self.upBody,
            smooth_segmentation= self.smooth,
            min_detection_confidence= self.detectionCon,
            min_tracking_confidence= self.trackCon
        )



    def findPose(self, img, draw=True):

        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        self.results = self.pose.process(imgRGB)

        if self.results.pose_landmarks:
            if draw:
                self.mpDraw.draw_landmarks(img, self.results.pose_landmarks, self.mpPose.POSE_CONNECTIONS)
        return img


    def findPosition(self, img, draw=True):
        lmList = []
        if self.results.pose_landmarks:
            for id, lm in enumerate(self.results.pose_landmarks.landmark):
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                lmList.append([id, cx, cy])
                if draw:
                    cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)
        return lmList


def main():
	pTime = 0
	cTime = 0
	picam2 = Picamera2()
	picam2.configure(picam2.create_preview_configuration(raw={"size": (1640, 1232)}, main={"format": 'RGB888', "size": (640, 480)}))
	picam2.start()
	time.sleep(2)
	detector = poseDetector()
	while True:
		img = picam2.capture_array()
		img = detector.findPose(img)
		lmList = detector.findPosition(img)
		if len(lmList) != 0:
			print(lmList)
		cTime = time.time()
		fps = 1 / (cTime - pTime)
		pTime = cTime

		cv2.putText(img, str(int(fps)), (10, 70), cv2.FONT_HERSHEY_PLAIN, 3, (0, 255, 0), 2)

		cv2.imshow("Output", img)
		if cv2.waitKey(1) & 0xFF == ord('q'):
			break

if __name__ == "__main__":
    main()
```

```python

import cv2
import numpy as np
from picamera2 import Picamera2
from PIL import Image
import PoseModule as pose
import time
import motors_command as ut


ut.init_gpio()


tolerance = 60
x_deviation = 0
y_max = 0

# -----initialise motor speed-----------------------------------

import RPi.GPIO as GPIO

GPIO.setmode(GPIO.BCM)  # choose BCM numbering scheme

GPIO.setup(20, GPIO.OUT)  # set GPIO 20 as output pin
GPIO.setup(21, GPIO.OUT)  # set GPIO 21 as output pin

pin20 = GPIO.PWM(20, 100)  # create object pin20 for PWM on port 20 at 100 Hertz
pin21 = GPIO.PWM(21, 100)  # create object pin21 for PWM on port 21 at 100 Hertz

val = 100  # maximum speed
pin20.start(val)  # start pin20 on 0 percent duty cycle (off)
pin21.start(val)  # start pin21 on 0 percent duty cycle (off)

print("speed set to: ", val)


# ------------------------------------------

def track_object(img, lmList):
    # global delay
    global x_deviation, y_max, tolerance

    h, w, c = img.shape
    print(h, w)

    flag = 0

    if len(lmList) != 0:
        if lmList[11][1] and lmList[11][2] and lmList[12][1] and lmList[12][2]:
            x_min, y_min, x_max, y_max = lmList[11][1], lmList[11][2], lmList[12][1], lmList[12][2]

            x_diff = x_max - x_min
            y_diff = y_max - y_min

            obj_x_center = x_min + (x_diff // 2)
            obj_x_center = round(obj_x_center, 3)

            obj_y_center = y_min + (y_diff // 2)
            obj_y_center = round(obj_y_center, 3)

            cv2.circle(img, (obj_x_center, obj_y_center), 15, (255, 0, 255), cv2.FILLED)
            flag = 1


    # print(x_min, y_min, x_max, y_max)
    if (flag == 0):
        print("selected object no present")
        return


    x_deviation = round(0.5 - obj_x_center, 3)
    y_max = round(y_max, 3)

    print("{", x_deviation, y_max, "}")

    move_robot(img, h, w, c)



def move_robot(img,h, w, c):
	global x_deviation, y_max, tolerance


	if (abs(x_deviation) > ((w//2)-tolerance) and abs(x_deviation) < ((w//2)+tolerance) ) :
		if (y_max < (h//2)):
			ut.stop()
			print("reached person...........")
			cv2.putText(img, "Forward", (450, 50), cv2.FONT_HERSHEY_PLAIN,
                        2, (0, 255, 0), 2)

		else:
			ut.forward()
			print("moving robot ...FORWARD....!!!!!!!!!!!!!!")
			cv2.putText(img, "Forward", (450, 50), cv2.FONT_HERSHEY_PLAIN,
                        2, (0, 255, 0), 2)


	else:
		if (abs(x_deviation) <= ((w//2) + tolerance)):
			delay1 = get_delay(x_deviation)

			ut.left()
			time.sleep(delay1)
			ut.stop()
			print("moving robot ...Left....<<<<<<<<<<")
			cv2.putText(img, "Left", (450, 80), cv2.FONT_HERSHEY_PLAIN,
                        2, (0, 255, 0), 2)

		if (abs(x_deviation) >= ((w//2) - tolerance)):
			delay1 = get_delay(x_deviation)

			ut.right()
			time.sleep(delay1)
			ut.stop()
			print("moving robot ...Right....>>>>>>>>")
			cv2.putText(img, "Right", (450, 80), cv2.FONT_HERSHEY_PLAIN,
                            2, (0, 255, 0), 2)


def get_delay(deviation):
    deviation = abs(deviation)

    if (deviation >= 0.4):
        d = 0.080
    elif (deviation >= 0.35 and deviation < 0.40):
        d = 0.060
    elif (deviation >= 0.20 and deviation < 0.35):
        d = 0.050
    else:
        d = 0.040

    return d


def main():
	pTime = 0
	cTime = 0
	picam2 = Picamera2()
	picam2.configure(picam2.create_preview_configuration(raw={"size": (1640, 1232)}, main={"format": 'RGB888', "size": (640, 480)}))
	picam2.start()
	time.sleep(2)
	detector = pose.poseDetector(detectionCon=0.75)
	while True:
		img = picam2.capture_array()
		img = cv2.rotate(img, cv2.ROTATE_180)
		img = detector.findPose(img)
		lmList = detector.findPosition(img)

		h, w, c = img.shape

		track_object(img, lmList)

		cv2.line(img, ((w//2) - 60 , 0), ((w//2) - 60, h), (0, 255, 0), 1)
		cv2.line(img, ((w//2) + 60 , 0), ((w//2) + 60, h), (0, 255, 0), 1)
		cv2.line(img, (0, h//2), (w, h//2), (0, 255, 0), 1)

		cTime = time.time()
		fps = 1 / (cTime - pTime)
		pTime = cTime

		cv2.putText(img, str(int(fps)), (10, 70), cv2.FONT_HERSHEY_PLAIN, 3, (0, 255, 0), 2)
		cv2.imshow("Output", img)
		if cv2.waitKey(1) & 0xFF == ord('q'):
			break

if __name__ == "__main__":
    main()
```
